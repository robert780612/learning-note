# Newton's method

Goal: Root finding

## Theorem

By Taylor expansion 

$$f(x) = f(x_0) + f'(x_0)(x-x_0)$$

Suppose f(x)=0, then

$$0=f(x_0)+f'(x_0)(x-x_0)$$

$$x_1=x_0-\frac{f(x_0)}{f'(x_0)}$$

So the process is repeated as

$$x_n=x_{n-1}-\frac{f(x_{n-1})}{f'(x_{n-1})}$$

## Example 

Find root for

$$y(x) = x^3 - x + 20e^x + 1$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def y(x):
    return x**3 - x + 20*np.exp(x) + 1

x = np.linspace(-3,3)
plt.plot(x, y(x))
plt.plot([-3, 3], [0, 0], 'r--')
plt.title("$x^3-x+20e^x+1$")
plt.show()
```

### Initial

First calculate y'

```python
def dy(x):
    return 3*x**2 - 1 + 20*np.exp(x)
```

Choose a $x_0$ (large slope is converge quickly)

```python
np.random.seed(123)
x0 = 3 - np.random.rand(1)
x1 = x0 - y(x0)/dy(x0)
print("x0=", x0)
print("x1=", x1)

plt.figure()
plt.plot(x, y(x))
plt.plot([-3, 3], [0, 0], 'r--')
plt.plot([x0, x1], [y(x0), 0], 'o--')
plt.text(x0-0.8, y(x0), '$f(x_0)$', fontsize=14)
plt.title("First iteration")
plt.show()
```

### Second iteration

```python
x2 = x1 - y(x1)/dy(x1)

plt.figure()
plt.plot(x, y(x))
plt.plot([-3, 3], [0, 0], 'r--')
plt.plot([x0, x1], [y(x0), 0], 'o--')
plt.text(x0-0.8, y(x0), '$f(x_0)$', fontsize=14)
plt.plot([x1, x2], [y(x1), 0], 'o--')
plt.text(x1-0.8, y(x1), '$f(x_1)$', fontsize=14)
plt.title("Second iteration")
plt.show()
```

### Third iteration

```python
x3 = x2 - y(x2)/dy(x2)

plt.figure()
plt.plot(x, y(x))
plt.plot([-3, 3], [0, 0], 'r--')
plt.plot([x0, x1], [y(x0), 0], 'o--')
plt.text(x0-0.8, y(x0), '$f(x_0)$', fontsize=14)
plt.plot([x1, x2], [y(x1), 0], 'o--')
plt.text(x1-0.8, y(x1), '$f(x_1)$', fontsize=14)
plt.plot([x2, x3], [y(x2), 0], 'o--')
plt.text(x2-0.8, y(x2), '$f(x_2)$', fontsize=14)
plt.title("Third iteration")
plt.show()
```

Obviously, $f(x_0)>f(x_1)>f(x_2)$, we can get the root of $y$ by more iteration

### More iteration

```python
def newtonMethod(y, dy, x, tor):
    for i in range(100):
        if y(x) < tor:
            break
        x = x - y(x)/dy(x)
    return x

xn = newtonMethod(y, dy, x0, 0.01)
print("xn=", xn, "y(xn)=", y(xn))

plt.figure()
plt.plot(x, y(x))
plt.plot([-3, 3], [0, 0], 'r--')
# x0
plt.plot([x0, x1], [y(x0), 0], 'o--')
plt.text(x0-0.8, y(x0), '$f(x_0)$', fontsize=14)
# x1
plt.plot([x1, x2], [y(x1), 0], 'o--')
plt.text(x1-0.8, y(x1), '$f(x_1)$', fontsize=14)
# x2
plt.plot([x2, x3], [y(x2), 0], 'o--')
plt.text(x2-0.8, y(x2), '$f(x_2)$', fontsize=14)
# xn
plt.plot([xn], [y(xn)], 'or')
plt.text(xn-0.8, y(xn)+10, '$f(x_n)$', fontsize=14)

plt.title("Final iteration")
plt.show()
```
